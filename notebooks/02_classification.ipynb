
# **1. Import Library**

```python
import pandas as pd
import joblib

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import classification_report
```

# **2. Load Dataset Hasil Clustering**

```python
df = pd.read_csv('../dataset/data_clustering_inverse.csv')
df.head()
```

# **Feature Encoding (OHE)**

```python
# Ambil kolom kategorik
categorical_cols = list(df.select_dtypes(include=['object']).columns)

# OHE dengan pd.ge_dummies
df_encoded = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True
)

df_encoded.head()
```

# **Data Splitting**

```python
# Buat 'x' sebagai prediktor
x = df_encoded.drop('Target', axis=1)

# Buat 'y' sebagai variabel target
y = df_encoded['Target']

# Split data
x_train, y_train, x_test, y_test = train_test_split(
    x,
    y,
    test_size = 0.2,
    random_state = 42,
    stratify = y
)

print('Jumlah total data :', len(x))
print('Jumlah data train :', len(x_train))
print('Jumlah data test  :', len(x_test))
```

# **3. Membangun Model Klasifikasi**

```python
# Gunakan algoritma DecisionTree
decision_tree_model = DecisionTreeClassifier(random_state=42)

# 2. Latih (fit) model dengan data training (X_train dan y_train)
decision_tree_model.fit(x_train, y_train)
```
